## Part 1 — Mean: Finding the Center

### Concept

The **mean** (average) tells you the single value that best represents the whole group.

$$\bar{x} = \frac{\text{sum of all values}}{\text{number of values}} = \frac{\sum x_i}{n}$$

### Intuition

Imagine balancing a see-saw. The mean is the exact point where it balances. Values larger than the mean pull one side; smaller values pull the other.

````python
# ── Dataset 1: Test Scores ────────────────────────────────────────────────────
test_scores = np.array([65, 78, 82, 71, 88, 74, 91, 68])
n = len(test_scores)

print('=== DATASET 1: Test Scores ===')
print(f'  Values : {test_scores}')
print(f'  Step 1  — sum of all values : {test_scores.sum()}')
print(f'  Step 2  — divide by n ({n})  : {test_scores.sum()} / {n} = {test_scores.sum()/n:.2f}')
mean_scores = test_scores.mean()
print(f'  Mean   : {mean_scores:.2f}')

# ── Dataset 2: Summer and Winter Temperatures ─────────────────────────────────
temp_summer = np.array([28, 32, 35, 31, 29, 33, 36, 30])
temp_winter = np.array([2, -1, 5, 3, 0, 4, 1, -2])

print('\n=== DATASET 2: Daily Temperatures (°C) ===')
for label, data in [('Summer', temp_summer), ('Winter', temp_winter)]:
    print(f'  {label}: values={data}, sum={data.sum()}, n={len(data)}, mean={data.mean():.2f}')

# ── Dataset 3: Player Heights ─────────────────────────────────────────────────
height_basketball = np.array([195, 201, 198, 210, 188, 205])
height_soccer     = np.array([170, 175, 168, 180, 172, 177])

print('\n=== DATASET 3: Heights (cm) ===')
for label, data in [('Basketball', height_basketball), ('Soccer', height_soccer)]:
    print(f'  {label}: values={data}, mean={data.mean():.2f}')

# ── Dataset 4: House Prices ───────────────────────────────────────────────────
price_A = np.array([250, 310, 280, 320, 270, 295])
price_B = np.array([450, 520, 480, 510, 495, 530])

print('\n=== DATASET 4: House Prices ($k) ===')
for label, data in [('Neighborhood A', price_A), ('Neighborhood B', price_B)]:
    print(f'  {label}: mean = ${data.mean():.1f}k')
````

````python
# ── Visualization: Bar charts with mean line ─────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(14, 9))
fig.suptitle('Dataset Values with Mean Line', fontsize=14, fontweight='bold')

datasets = [
    ('Test Scores',          test_scores,         'Student',     'Score',       '#4C72B0'),
    ('Summer Temperatures',  temp_summer,          'Day',         'Temp (°C)',   '#DD8452'),
    ('Basketball Heights',   height_basketball,    'Player',      'Height (cm)', '#55A868'),
    ('House Prices — Nbhd A', price_A,             'House',       'Price ($k)',  '#C44E52'),
]

for ax, (title, data, xlabel, ylabel, color) in zip(axes.flat, datasets):
    xs = np.arange(1, len(data) + 1)
    ax.bar(xs, data, color=color, alpha=0.75, edgecolor='white', linewidth=0.8)
    ax.axhline(data.mean(), color='black', linewidth=2.0, linestyle='--',
               label=f'Mean = {data.mean():.1f}')
    ax.set_title(title, fontweight='bold')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_xticks(xs)
    ax.legend()

plt.tight_layout()
plt.show()
````

### Key Insight

The mean summarizes an entire group with one number. Notice that the mean of basketball players (~199 cm) vs soccer players (~174 cm) immediately tells us basketball players are substantially taller — no need to compare each individual.

## Part 2 — Standard Deviation: How Spread Out Is the Data?

### Concept

Two datasets can have the same mean but look completely different. Standard deviation measures how far values typically stray from the mean.

$$\sigma = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n}}$$

### The 5 Steps

1. Subtract the mean from each value → **deviations**
2. Square each deviation → eliminate negatives, amplify large gaps
3. Average the squared deviations → **variance**
4. Take the square root → bring units back
5. Interpret: small σ = tight cluster, large σ = wide spread

````python
def std_step_by_step(data, label):
    """Print all 5 steps of standard deviation calculation."""
    print(f'\n=== {label} ===')
    print(f'  Data   : {data}')

    # Step 1 — subtract mean
    mean = data.mean()
    deviations = data - mean
    print(f'  Step 1  — subtract mean ({mean:.2f}):')
    print(f'            {np.round(deviations, 2)}')

    # Step 2 — square deviations
    sq_dev = deviations ** 2
    print(f'  Step 2  — square each deviation:')
    print(f'            {np.round(sq_dev, 2)}')

    # Step 3 — average (variance)
    variance = sq_dev.mean()
    print(f'  Step 3  — average squared deviations (variance) = {variance:.4f}')

    # Step 4 — square root
    std = np.sqrt(variance)
    print(f'  Step 4  — square root of variance = {std:.4f}')

    # Step 5 — interpret
    print(f'  Step 5  — std dev = {std:.2f}  '
          f'(values typically within {mean:.1f} ± {std:.1f})')
    return mean, std


# Run all four datasets
mean_ts, std_ts  = std_step_by_step(test_scores,         'Test Scores')
mean_su, std_su  = std_step_by_step(temp_summer,         'Summer Temperatures (°C)')
mean_wi, std_wi  = std_step_by_step(temp_winter,         'Winter Temperatures (°C)')
mean_bk, std_bk  = std_step_by_step(height_basketball,   'Basketball Player Heights (cm)')
mean_sc, std_sc  = std_step_by_step(height_soccer,       'Soccer Player Heights (cm)')
mean_pA, std_pA  = std_step_by_step(price_A,             'House Prices — Neighborhood A ($k)')
mean_pB, std_pB  = std_step_by_step(price_B,             'House Prices — Neighborhood B ($k)')
````

````python
# ── Visualization: Compare two groups side by side (histograms) ───────────────
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Comparing Groups — Same Scale Reveals Spread Differences',
             fontsize=13, fontweight='bold')

# Panel 1 — Summer vs Winter
ax = axes[0]
all_temps = np.concatenate([temp_summer, temp_winter])
bins = np.arange(all_temps.min() - 2, all_temps.max() + 4, 2)
ax.hist(temp_summer, bins=bins, alpha=0.65, color='#DD8452', label=f'Summer μ={mean_su:.1f} σ={std_su:.1f}')
ax.hist(temp_winter, bins=bins, alpha=0.65, color='#4C72B0', label=f'Winter μ={mean_wi:.1f} σ={std_wi:.1f}')
ax.set_title('Temperature (°C)')
ax.set_xlabel('Temperature')
ax.set_ylabel('Frequency')
ax.legend()

# Panel 2 — Basketball vs Soccer heights
ax = axes[1]
all_h = np.concatenate([height_basketball, height_soccer])
bins_h = np.arange(all_h.min() - 5, all_h.max() + 8, 5)
ax.hist(height_basketball, bins=bins_h, alpha=0.65, color='#55A868',
        label=f'Basketball μ={mean_bk:.1f} σ={std_bk:.1f}')
ax.hist(height_soccer,     bins=bins_h, alpha=0.65, color='#C44E52',
        label=f'Soccer μ={mean_sc:.1f} σ={std_sc:.1f}')
ax.set_title('Player Heights (cm)')
ax.set_xlabel('Height (cm)')
ax.legend()

# Panel 3 — House prices A vs B
ax = axes[2]
all_p = np.concatenate([price_A, price_B])
bins_p = np.arange(all_p.min() - 20, all_p.max() + 40, 20)
ax.hist(price_A, bins=bins_p, alpha=0.65, color='#8172B2',
        label=f'Nbhd A μ={mean_pA:.0f} σ={std_pA:.0f}')
ax.hist(price_B, bins=bins_p, alpha=0.65, color='#64B5CD',
        label=f'Nbhd B μ={mean_pB:.0f} σ={std_pB:.0f}')
ax.set_title('House Prices ($k)')
ax.set_xlabel('Price ($k)')
ax.legend()

plt.tight_layout()
plt.show()
````

### Key Insight

Both summer and winter data have similar spread (σ ≈ 2.5–2.8), but their means are ~18 °C apart. The basketball group is taller AND more variable than the soccer group. Neighborhood B has higher prices AND more variation — the standard deviation captures both center and spread of each group independently.

## Part 3 — Z-Score: The Universal Ruler

### Concept

A z-score answers: **"How unusual is this value compared to the rest of the group?"**

$$z_i = \frac{x_i - \bar{x}}{\sigma}$$

- z = 0 → exactly average
- z = +1 → one standard deviation above average
- z = -2 → two standard deviations below average
- |z| > 3 → very unusual (possible outlier)

### Why this is powerful

Z-scores let you compare values on completely different scales. Is a score of 91 in a class the same as being 6'10" on a basketball team? Z-scores give you a single answer.

````python
def zscore(data):
    """Compute z-scores: (x - mean) / std"""
    return (data - data.mean()) / data.std()


# Compute z-scores for all datasets
z_scores  = zscore(test_scores)
z_summer  = zscore(temp_summer)
z_winter  = zscore(temp_winter)
z_basket  = zscore(height_basketball)
z_soccer  = zscore(height_soccer)
z_priceA  = zscore(price_A)
z_priceB  = zscore(price_B)

# Detailed walkthrough for test scores
print('=== Z-SCORES: Test Scores (detailed walkthrough) ===')
print(f'  Mean = {test_scores.mean():.2f},  Std = {test_scores.std():.2f}')
print(f'  {"Score":>6}  {"z-score":>8}  Interpretation')
print(f'  {"-"*45}')
for score, z in zip(test_scores, z_scores):
    interpretation = (
        'exactly average' if abs(z) < 0.1 else
        f'{abs(z):.2f} std {"above" if z > 0 else "below"} average'
    )
    print(f'  {score:>6}  {z:>+8.2f}  {interpretation}')

# Summary for all datasets
print('\n=== Z-SCORE SUMMARIES ===')
for label, data, z in [
        ('Summer temp',   temp_summer,        z_summer),
        ('Winter temp',   temp_winter,        z_winter),
        ('Basketball ht', height_basketball,  z_basket),
        ('Soccer ht',     height_soccer,      z_soccer),
        ('Prices A',      price_A,            z_priceA),
        ('Prices B',      price_B,            z_priceB),
    ]:
    print(f'  {label:15s} -> z-scores: {np.round(z, 2)}')
````

````python
# ── Visualization: Z-score comparison across datasets ────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(14, 9))
fig.suptitle('Z-Scores: Everything on the Same Scale (σ units from mean)',
             fontsize=13, fontweight='bold')

plot_data = [
    ('Test Scores',         np.arange(1, 9), [z_scores], ['#4C72B0'], ['Scores']),
    ('Temperatures',        np.arange(1, 9), [z_summer, z_winter], ['#DD8452','#4C72B0'], ['Summer','Winter']),
    ('Player Heights',      np.arange(1, 7), [z_basket, z_soccer], ['#55A868','#C44E52'], ['Basketball','Soccer']),
    ('House Prices',        np.arange(1, 7), [z_priceA, z_priceB], ['#8172B2','#64B5CD'], ['Nbhd A','Nbhd B']),
]

for ax, (title, xs, z_list, colors, labels) in zip(axes.flat, plot_data):
    for z, color, label in zip(z_list, colors, labels):
        ax.plot(xs, z, 'o-', color=color, linewidth=2, markersize=7, label=label)
    ax.axhline(0,   color='black',  linewidth=1.0, linestyle='--', alpha=0.6, label='Mean (z=0)')
    ax.axhline( 1,  color='gray',   linewidth=0.7, linestyle=':',  alpha=0.5)
    ax.axhline(-1,  color='gray',   linewidth=0.7, linestyle=':',  alpha=0.5)
    ax.set_title(title, fontweight='bold')
    ax.set_xlabel('Index')
    ax.set_ylabel('Z-Score (σ units)')
    ax.set_ylim(-3, 3)
    ax.legend(fontsize=9)

plt.tight_layout()
plt.show()

print('Dotted lines at ±1σ — values beyond those lines are somewhat unusual.')
````

### Key Insight

After computing z-scores, every dataset lives on the same scale (roughly -2 to +2). The player with the highest z-score in basketball is the most unusually tall *relative to other basketball players*, not in absolute centimeters. This is crucial for ML: your model can now compare test scores and house prices on equal footing.

## Part 4 — Why Normalization Matters in Machine Learning

### The Problem

Machine learning models learn by adjusting internal values called **weights** based on errors. If one feature (e.g., salary = 80,000) is on a much larger scale than another (e.g., age = 35), the model spends most of its effort adjusting for the large-scale feature and barely notices the small-scale one.

This is the **gradient magnitude problem**:

- Gradient for salary ≈ 80,000 × (error) → huge update
- Gradient for age ≈ 35 × (error) → tiny update

The model misbehaves because the features are on different scales.

````python
# ── Simulate gradient magnitudes before and after normalization ───────────────
np.random.seed(42)

# Raw features: age (20–60) and salary (30k–110k)
ages    = np.array([25, 32, 45, 28, 55, 38, 41, 22, 60, 34])
salaries = np.array([35000, 52000, 78000, 41000, 95000,
                     60000, 68000, 30000, 110000, 55000])

# A simplified stand-in for gradient: feature × error
# Pretend the error for this step is 0.1
error = 0.1

grad_age_raw    = ages    * error
grad_salary_raw = salaries * error

print('=== BEFORE NORMALIZATION ===')
print(f'  Gradient magnitude for age    : min={grad_age_raw.min():.1f},  '
      f'max={grad_age_raw.max():.1f},  mean={grad_age_raw.mean():.1f}')
print(f'  Gradient magnitude for salary : min={grad_salary_raw.min():.0f}, '
      f'max={grad_salary_raw.max():.0f}, mean={grad_salary_raw.mean():.0f}')
print(f'  Ratio salary/age gradient     : {grad_salary_raw.mean()/grad_age_raw.mean():.0f}x larger!')

# Normalize both features using z-score
z_ages    = zscore(ages)
z_salaries = zscore(salaries)

grad_age_norm    = z_ages    * error
grad_salary_norm = z_salaries * error

print('\n=== AFTER Z-SCORE NORMALIZATION ===')
print(f'  Gradient magnitude for age    : min={grad_age_norm.min():.4f}, '
      f'max={grad_age_norm.max():.4f}, mean≈{grad_age_norm.mean():.4f}')
print(f'  Gradient magnitude for salary : min={grad_salary_norm.min():.4f}, '
      f'max={grad_salary_norm.max():.4f}, mean≈{grad_salary_norm.mean():.4f}')
print('  Now both features contribute equally to learning!')
````

````python
# ── Visualization: Raw vs normalized feature distributions ───────────────────
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Normalization Effect: Raw Features vs Z-Score Normalized',
             fontsize=13, fontweight='bold')

# Row 0 — raw features (completely different scales)
ax = axes[0, 0]
ax.bar(np.arange(len(ages)), ages, color='#4C72B0', alpha=0.8)
ax.set_title('Raw Age Values (20–60)', fontweight='bold')
ax.set_ylabel('Age (years)')
ax.set_xlabel('Person index')

ax = axes[0, 1]
ax.bar(np.arange(len(salaries)), salaries / 1000, color='#DD8452', alpha=0.8)
ax.set_title('Raw Salary Values (30k–110k)', fontweight='bold')
ax.set_ylabel('Salary ($k)')
ax.set_xlabel('Person index')

# Row 1 — normalized (same scale!)
ax = axes[1, 0]
ax.bar(np.arange(len(z_ages)), z_ages, color='#4C72B0', alpha=0.8)
ax.axhline(0, color='black', linewidth=1.2, linestyle='--')
ax.set_title('Normalized Age (z-score)', fontweight='bold')
ax.set_ylabel('Z-Score')
ax.set_xlabel('Person index')
ax.set_ylim(-2.5, 2.5)

ax = axes[1, 1]
ax.bar(np.arange(len(z_salaries)), z_salaries, color='#DD8452', alpha=0.8)
ax.axhline(0, color='black', linewidth=1.2, linestyle='--')
ax.set_title('Normalized Salary (z-score)', fontweight='bold')
ax.set_ylabel('Z-Score')
ax.set_xlabel('Person index')
ax.set_ylim(-2.5, 2.5)

plt.tight_layout()
plt.show()
print('Bottom row: both features now live in the same ±2 range — the model treats them fairly.')
````

````python
# ── Visualization: Gradient magnitude comparison bar chart ────────────────────
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle('Gradient Magnitudes: Why Unscaled Features Break Learning',
             fontsize=13, fontweight='bold')

# Before normalization
labels_feat = ['Age\n(raw)', 'Salary\n(raw, /1000)']
raw_means   = [grad_age_raw.mean(), grad_salary_raw.mean() / 1000]
bars = ax1.bar(labels_feat, raw_means, color=['#4C72B0', '#DD8452'], alpha=0.8, width=0.4)
ax1.set_title('Before Normalization', fontweight='bold')
ax1.set_ylabel('Mean Gradient Magnitude')
ax1.bar_label(bars, fmt='%.1f', padding=3)
ax1.set_ylim(0, max(raw_means) * 1.3)

# After normalization
labels_feat2  = ['Age\n(z-score)', 'Salary\n(z-score)']
norm_means    = [abs(grad_age_norm).mean(), abs(grad_salary_norm).mean()]
bars2 = ax2.bar(labels_feat2, norm_means, color=['#4C72B0', '#DD8452'], alpha=0.8, width=0.4)
ax2.set_title('After Z-Score Normalization', fontweight='bold')
ax2.set_ylabel('Mean Gradient Magnitude')
ax2.bar_label(bars2, fmt='%.4f', padding=3)
ax2.set_ylim(0, max(norm_means) * 1.5)

plt.tight_layout()
plt.show()
````

### Key Insight

Before normalization, the salary gradient is ~1,600x larger than the age gradient. The model would update its salary weight aggressively while almost ignoring age — even if age is actually an important predictor. After z-score normalization, both gradients are tiny and comparable. The model can learn from both features equally.

## Part 5 — The Bell Curve: The Shape of Natural Data

### Concept

Many real-world quantities follow a **normal distribution** (bell curve). When they do:

- ~68% of values fall within 1 standard deviation of the mean
- ~95% fall within 2 standard deviations
- ~99.7% fall within 3 standard deviations

This is called the **68-95-99.7 rule**. Z-scores tell you exactly where on this bell curve a value lands.

````python
# ── Draw the normal distribution and shade ±1σ, ±2σ, ±3σ regions ─────────────
from matplotlib.patches import Patch

def normal_pdf(x, mu=0, sigma=1):
    """Standard normal probability density function."""
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

x = np.linspace(-4, 4, 500)
y = normal_pdf(x)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('The Normal Distribution (Bell Curve)', fontsize=13, fontweight='bold')

# Panel 1 — the 68-95-99.7 rule
ax = axes[0]
ax.plot(x, y, color='black', linewidth=2.5)
fill_specs = [
    (3,  '#2196F3', 0.20, '99.7% within ±3σ'),
    (2,  '#42A5F5', 0.35, '95% within ±2σ'),
    (1,  '#90CAF9', 0.55, '68% within ±1σ'),
]
for sigma_band, color, alpha, label in fill_specs:
    mask = (x >= -sigma_band) & (x <= sigma_band)
    ax.fill_between(x, y, where=mask, color=color, alpha=alpha, label=label)

for sigma_val, label in [(-3,'−3σ'),(-2,'−2σ'),(-1,'−1σ'),(0,'Mean'),(1,'+1σ'),(2,'+2σ'),(3,'+3σ')]:
    ax.axvline(sigma_val, color='gray', linewidth=0.8, linestyle=':')
    ax.text(sigma_val, -0.012, label, ha='center', fontsize=8, color='gray')

ax.set_title('68-95-99.7 Rule')
ax.set_xlabel('Z-Score (standard deviations from mean)')
ax.set_ylabel('Probability Density')
ax.legend(loc='upper right', fontsize=9)
ax.set_ylim(-0.03, 0.44)

# Panel 2 — Where do our test scores fall?
ax = axes[1]
mu_ts, sigma_ts = test_scores.mean(), test_scores.std()
x2 = np.linspace(mu_ts - 4*sigma_ts, mu_ts + 4*sigma_ts, 500)
y2 = normal_pdf(x2, mu=mu_ts, sigma=sigma_ts)
ax.plot(x2, y2, color='#4C72B0', linewidth=2.5)

colors_pts = ['#C44E52' if abs(z) > 2 else '#DD8452' if abs(z) > 1 else '#55A868'
              for z in z_scores]
for score, z, color in zip(test_scores, z_scores, colors_pts):
    ax.axvline(score, color=color, linewidth=1.8, linestyle='--', alpha=0.8)
    ax.text(score, normal_pdf(score, mu_ts, sigma_ts) + 0.003,
            f'{score}\n(z={z:+.1f})', ha='center', fontsize=8)

ax.set_title(f'Test Scores on Their Bell Curve\n(μ={mu_ts:.1f}, σ={sigma_ts:.1f})',
             fontweight='bold')
ax.set_xlabel('Score')
ax.set_ylabel('Probability Density')

legend_elements = [
    Patch(facecolor='#55A868', label='|z| ≤ 1 (common)'),
    Patch(facecolor='#DD8452', label='1 < |z| ≤ 2 (uncommon)'),
    Patch(facecolor='#C44E52', label='|z| > 2 (rare)'),
]
ax.legend(handles=legend_elements, fontsize=9)

plt.tight_layout()
plt.show()
````

### Key Insight

The bell curve is not just a pretty shape — it is a model of how natural variation works. When you convert any measurement to z-scores, you are essentially asking *where on this standard bell curve does this value sit?* A z-score of +2 on a test means roughly the top 2.5% — that interpretation is the same whether you are measuring test scores, blood pressure, or house prices.

````python
# ══════════════════════════════════════════════════════════════════════════════
# ANSWER CELL — only look after you have tried it yourself!
# ══════════════════════════════════════════════════════════════════════════════

bp = np.array([120, 135, 118, 142, 128, 155, 122, 138])

# 1. Mean
bp_mean = bp.mean()
print(f'1. Mean = {bp.sum()} / {len(bp)} = {bp_mean:.2f} mmHg')

# 2. Std dev — 5 steps
print('\n2. Standard Deviation:')
deviations = bp - bp_mean
print(f'   Step 1 deviations      : {np.round(deviations, 2)}')
sq_dev = deviations ** 2
print(f'   Step 2 squared devs    : {np.round(sq_dev, 2)}')
variance = sq_dev.mean()
print(f'   Step 3 variance        : {variance:.4f}')
bp_std = np.sqrt(variance)
print(f'   Step 4 sqrt(variance)  : {bp_std:.4f}')
print(f'   Step 5 std dev         : {bp_std:.2f} mmHg')

# 3. Z-scores
bp_z = (bp - bp_mean) / bp_std
print('\n3. Z-scores:')
for patient, reading, z in zip(range(1, 9), bp, bp_z):
    print(f'   Patient {patient}: {reading} mmHg  →  z = {z:+.2f}')

# 4. Identify extremes
most_unusual_idx   = np.argmax(np.abs(bp_z))
closest_avg_idx    = np.argmin(np.abs(bp_z))
print(f'\n4. Most unusual (highest |z|): Patient {most_unusual_idx+1}, '
      f'reading={bp[most_unusual_idx]}, z={bp_z[most_unusual_idx]:+.2f}')
print(f'   Closest to average (z≈0) : Patient {closest_avg_idx+1}, '
      f'reading={bp[closest_avg_idx]}, z={bp_z[closest_avg_idx]:+.2f}')

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle('Blood Pressure Exercise — Answer', fontsize=13, fontweight='bold')

patients = np.arange(1, 9)
ax1.bar(patients, bp, color='#4C72B0', alpha=0.75, edgecolor='white')
ax1.axhline(bp_mean, color='red', linewidth=2, linestyle='--',
            label=f'Mean = {bp_mean:.1f}')
ax1.set_title('Raw Readings')
ax1.set_xlabel('Patient')
ax1.set_ylabel('BP (mmHg)')
ax1.set_xticks(patients)
ax1.legend()

bar_colors = ['#C44E52' if abs(z) > 1.5 else '#55A868' for z in bp_z]
bars = ax2.bar(patients, bp_z, color=bar_colors, alpha=0.8, edgecolor='white')
ax2.axhline(0, color='black', linewidth=1.5, linestyle='--')
ax2.axhline(1, color='gray',  linewidth=0.8, linestyle=':',  alpha=0.7)
ax2.axhline(-1, color='gray', linewidth=0.8, linestyle=':',  alpha=0.7)
ax2.set_title('Z-Scores (red = >1.5σ from mean)')
ax2.set_xlabel('Patient')
ax2.set_ylabel('Z-Score')
ax2.set_xticks(patients)
ax2.bar_label(bars, fmt='%.2f', padding=2, fontsize=9)

plt.tight_layout()
plt.show()
````
