# WEEK 1: Statistics Foundation
## Concept: Mean — The Center of Data

**Key Insight:** Before we can build any model, we need to understand our data.  
The **mean** tells us where the data is centered.

**Formula:**
$$\mu = \frac{\sum x_i}{n}$$

**Step-by-step:**
1. Add all values
2. Divide by the count

````python
# --- Exercise 1.1: Computing Mean ---
# A teacher records student test scores
scores = [65, 78, 82, 71, 88]

# Step-by-step by hand
total = sum(scores)
count = len(scores)
mean = total / count

print('Scores:', scores)
print(f'Sum: {total}')
print(f'Count: {count}')
print(f'Mean = {total} / {count} = {mean}')

# Verify with numpy
print(f'\nNumpy confirms: {np.mean(scores)}')

# Visualize
fig, ax = plt.subplots(figsize=(9, 4))
ax.bar(range(len(scores)), scores, color='steelblue', alpha=0.7, label='Scores')
ax.axhline(mean, color='red', linewidth=2, linestyle='--', label=f'Mean = {mean}')
ax.set_xticks(range(len(scores)))
ax.set_xticklabels([f'Student {i+1}' for i in range(len(scores))])
ax.set_ylabel('Score')
ax.set_title('Student Scores with Mean')
ax.legend()
plt.tight_layout()
plt.show()
print('\nCONCEPT: The mean is the center. Most scores are around 76.8.')
````

## Concept: Standard Deviation — The Spread of Data

Mean tells us the center. **Standard deviation** tells us how spread out the data is.

**Formula:**
$$\sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{n}}$$

**5 Steps:**
1. Subtract mean from each value  
2. Square each difference  
3. Average the squares (= **variance**)  
4. Take the square root  
5. Result = std dev

````python
# --- Exercise 1.2: Computing Standard Deviation ---
scores = [65, 78, 82, 71, 88]
mean = 76.8

print('Step 1: Subtract mean from each score')
deviations = [(x, x - mean) for x in scores]
for score, dev in deviations:
    print(f'  {score} - {mean} = {dev:.1f}')

print('\nStep 2: Square each deviation')
squared_devs = [(dev, dev**2) for _, dev in deviations]
for dev, sq in squared_devs:
    print(f'  ({dev:.1f})² = {sq:.2f}')

squares = [sq for _, sq in squared_devs]
variance = sum(squares) / len(squares)
std_dev = variance ** 0.5

print(f'\nStep 3: Variance = {sum(squares):.2f} / {len(squares)} = {variance:.2f}')
print(f'Step 4: Std Dev = √{variance:.2f} ≈ {std_dev:.2f}')
print(f'\nNumpy confirms: {np.std(scores):.2f}')

# Visualize spread
fig, ax = plt.subplots(figsize=(9, 4))
ax.scatter(scores, [0]*len(scores), s=150, color='steelblue', zorder=5, label='Scores')
ax.axvline(mean, color='red', linewidth=2, label=f'Mean = {mean}')
ax.axvspan(mean - std_dev, mean + std_dev, alpha=0.15, color='orange', label=f'±1 Std Dev ({std_dev:.1f})')
ax.set_yticks([])
ax.set_xlabel('Score')
ax.set_title('Score Distribution: Mean and Spread')
ax.legend()
plt.tight_layout()
plt.show()
print('\nCONCEPT: Typical distance from mean is ~8 points. Scores spread ±8 around 76.8.')
````

## Concept: Z-Score — Putting Everything on the Same Scale

A **z-score** tells us how many standard deviations a value is from the mean.

**Formula:**
$$z = \frac{x - \mu}{\sigma}$$

**Why it matters:** Z-scores always fall roughly in [-3, +3] regardless of the original units.  
This is how we compare apples to oranges (e.g., age vs. income).

````python
# --- Exercise 1.3: Computing Z-Scores ---
scores = [65, 78, 82, 71, 88]
mean = 76.8
std_dev = 8.08

print('Z-Score Formula: z = (x - mean) / std_dev\n')
z_scores = []
for score in scores:
    z = (score - mean) / std_dev
    z_scores.append(z)
    interpretation = 'above' if z > 0 else 'below'
    print(f'  Score {score}: z = ({score} - {mean}) / {std_dev} = {z:.2f}  →  {abs(z):.2f} std devs {interpretation} mean')

print(f'\nZ-scores: {[round(z, 2) for z in z_scores]}')
print(f'All in range [{min(z_scores):.2f}, {max(z_scores):.2f}] ✓')

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].bar(range(len(scores)), scores, color='steelblue', alpha=0.7)
axes[0].axhline(mean, color='red', linestyle='--', label=f'Mean={mean}')
axes[0].set_title('Original Scores')
axes[0].set_ylabel('Score')
axes[0].set_xticks(range(len(scores)))
axes[0].set_xticklabels([f'S{i+1}' for i in range(len(scores))])
axes[0].legend()

colors = ['green' if z >= 0 else 'red' for z in z_scores]
axes[1].bar(range(len(z_scores)), z_scores, color=colors, alpha=0.7)
axes[1].axhline(0, color='black', linewidth=1.5)
axes[1].axhline(1, color='orange', linestyle=':', alpha=0.7, label='+1 std dev')
axes[1].axhline(-1, color='orange', linestyle=':', alpha=0.7, label='-1 std dev')
axes[1].set_title('Z-Scores (Normalized)')
axes[1].set_ylabel('Z-Score')
axes[1].set_xticks(range(len(z_scores)))
axes[1].set_xticklabels([f'S{i+1}' for i in range(len(z_scores))])
axes[1].legend()

plt.suptitle('Raw Scores vs Z-Scores', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

## Concept: Why Normalization Matters

Without normalization, features with large values **dominate** features with small values.  
This breaks learning.

````python
# --- Exercise 1.4: The Normalization Problem ---
ages = [20, 30, 40, 50, 60]          # range: 40 units
incomes = [30000, 50000, 70000, 90000, 110000]  # range: 80,000 units!

print('WITHOUT normalization:')
print(f'  Age range:    {min(ages)} to {max(ages)} = {max(ages)-min(ages)} units')
print(f'  Income range: ${min(incomes):,} to ${max(incomes):,} = ${max(incomes)-min(incomes):,} units')
print(f'  Income is {(max(incomes)-min(incomes))/(max(ages)-min(ages)):.0f}× LARGER!')
print('  → Income gradient dominates, age is ignored!')

# Normalize both
def z_normalize(data):
    m = np.mean(data)
    s = np.std(data)
    return [(x - m) / s for x in data]

z_ages = z_normalize(ages)
z_incomes = z_normalize(incomes)

print('\nWITH normalization (z-scores):')
print(f'  Z-Ages:    {[round(z, 2) for z in z_ages]}')
print(f'  Z-Incomes: {[round(z, 2) for z in z_incomes]}')
print(f'  Both in range [{min(z_ages):.2f}, {max(z_ages):.2f}] — same scale! ✓')

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

x = range(5)
axes[0].plot(x, ages, 'o-', color='blue', label='Age', linewidth=2)
ax2 = axes[0].twinx()
ax2.plot(x, incomes, 's-', color='orange', label='Income', linewidth=2)
axes[0].set_title('RAW: Different Scales (Unfair!)', color='red')
axes[0].set_ylabel('Age (units)', color='blue')
ax2.set_ylabel('Income ($)', color='orange')
axes[0].set_xlabel('Person')
lines1, _ = axes[0].get_legend_handles_labels()
lines2, _ = ax2.get_legend_handles_labels()
axes[0].legend(lines1 + lines2, ['Age', 'Income'])

axes[1].plot(x, z_ages, 'o-', color='blue', label='Z-Age', linewidth=2)
axes[1].plot(x, z_incomes, 's-', color='orange', label='Z-Income', linewidth=2)
axes[1].axhline(0, color='black', linewidth=0.8)
axes[1].set_title('NORMALIZED: Same Scale (Fair!)', color='green')
axes[1].set_ylabel('Z-Score')
axes[1].set_xlabel('Person')
axes[1].legend()

plt.tight_layout()
plt.show()
print('\nKEY INSIGHT: Normalization prevents large features from bullying small ones.')
````

# WEEK 2: The Problem — Why Statistics Alone Aren't Enough
## Concept: Description vs Prediction

Statistics **describe** what happened in the past.  
A **model** predicts what will happen for a specific new case.

> "Knowing the average age doesn't tell us if **this specific person** should be approved."

````python
# --- Exercise 2.1: Statistics vs Prediction ---
print('=== LOAN APPROVAL SCENARIO ===')
print('Historical statistics:')
print('  80% of customers aged 40+ were approved')
print('  20% of customers aged under 40 were approved')
print()
print('New customer: Age 35')
print()
print('Question: Should we approve this customer?')
print()
print('What statistics says: "Young people are usually rejected"')
print('What statistics CANNOT say: "This specific person should be rejected"')
print()
print('Why? Because:')
print('  - 20% of young people ARE approved (maybe this is one of them)')
print('  - Approval depends on other features: income, credit score, job history...')
print('  - Statistics describe trends, not individuals')
print()
print('We need a MODEL that:')
print('  → Uses ALL features simultaneously')
print('  → Computes a probability for THIS person')
print('  → Makes a specific yes/no decision')

# Visualize why statistics fail
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Statistics view
ages_above_40 = [0.8, 0.2]
ages_below_40 = [0.2, 0.8]
axes[0].bar(['Approved', 'Rejected'], ages_above_40, color=['green', 'red'], alpha=0.7, label='Age 40+')
axes[0].bar(['Approved', 'Rejected'], ages_below_40, color=['lightgreen', 'lightcoral'], alpha=0.5, 
            bottom=[0, 0], label='Age <40', hatch='//')
axes[0].set_title('Statistics: Group-level info only')
axes[0].set_ylabel('Proportion')
axes[0].legend()

# Model view
np.random.seed(42)
ages = np.random.uniform(20, 70, 50)
incomes = np.random.uniform(20, 100, 50)
approved = ((0.3 * (ages - 45) / 15 + 0.5 * (incomes - 60) / 30) > 0).astype(int)

colors = ['green' if a == 1 else 'red' for a in approved]
axes[1].scatter(ages, incomes, c=colors, alpha=0.7, s=80)
axes[1].scatter([35], [55], c='gold', s=200, marker='*', zorder=5, label='New customer (35, $55k)')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Income ($k)')
axes[1].set_title('Model: Individual prediction with all features')
green_patch = mpatches.Patch(color='green', label='Approved')
red_patch = mpatches.Patch(color='red', label='Rejected')
axes[1].legend(handles=[green_patch, red_patch, plt.Line2D([0], [0], marker='*', color='gold', markersize=10, label='New customer')])

plt.tight_layout()
plt.show()
````

## Concept: Error — The Distance Between Prediction and Reality

A model makes a prediction. The **error** measures how wrong it was.

**Squared Error for one example:**
$$L = (\hat{y} - y)^2$$

**Mean Squared Error (MSE) over many examples:**
$$MSE = \frac{1}{n}\sum_{i=1}^n (\hat{y}_i - y_i)^2$$

Goal of learning: **minimize error**.

````python
# --- Exercise 2.2: Understanding Error ---
predictions = [0.2, 0.8, 0.6, 0.4, 0.9]
actuals     = [0,   1,   1,   0,   1  ]

print('Prediction | Actual | Error = (pred - actual)² | Interpretation')
print('-' * 70)
errors = []
for pred, actual in zip(predictions, actuals):
    error = (pred - actual) ** 2
    errors.append(error)
    quality = 'good' if error < 0.1 else ('ok' if error < 0.2 else 'bad')
    print(f'  {pred:.1f}       |   {actual}    | {error:.4f}                  | {quality}')

total_error = sum(errors)
mse = total_error / len(errors)
print(f'\nTotal error: {total_error:.4f}')
print(f'Mean Squared Error (MSE): {mse:.4f}')
print('\nCONCEPT: Smaller MSE = better predictions. Our goal is to MINIMIZE this.')

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

x = range(len(predictions))
axes[0].bar(x, actuals, color='green', alpha=0.4, label='Actual', width=0.4, align='edge')
axes[0].bar([i - 0.4 for i in x], predictions, color='blue', alpha=0.4, label='Prediction', width=0.4, align='edge')
for i, (p, a) in enumerate(zip(predictions, actuals)):
    axes[0].plot([i, i], [p, a], 'r-', linewidth=2, alpha=0.8)
axes[0].set_title('Predictions vs Actuals (red = error)')
axes[0].set_ylabel('Value')
axes[0].set_xticks(x)
axes[0].set_xticklabels([f'Ex {i+1}' for i in x])
axes[0].legend()

axes[1].bar(x, errors, color=['green' if e < 0.1 else 'orange' if e < 0.2 else 'red' for e in errors], alpha=0.8)
axes[1].axhline(mse, color='black', linestyle='--', label=f'MSE = {mse:.3f}')
axes[1].set_title('Squared Errors per Example')
axes[1].set_ylabel('Squared Error')
axes[1].set_xticks(x)
axes[1].set_xticklabels([f'Ex {i+1}' for i in x])
axes[1].legend()

plt.tight_layout()
plt.show()
````

# WEEK 3: The Solution — Linear Model
## Concept: ŷ = w×z + b

A **linear model** multiplies each feature by a **weight**, sums everything, then adds a **bias**.

$$\hat{y} = w_1 z_1 + w_2 z_2 + \ldots + w_n z_n + b$$

- **Weights (w):** How much each feature matters
- **Bias (b):** The baseline prediction
- **z:** Normalized (z-score) features

The model's job: find weights that minimize error.

````python
# --- Exercise 3.1: Making Predictions ---
# Model: ŷ = 0.3×z_age + 0.4×z_income + 0.2×z_credit + 0.1
w_age    = 0.3
w_income = 0.4
w_credit = 0.2
b        = 0.1

# New customer (already normalized)
z_age    =  0.5   # 0.5 std devs above average age
z_income = -0.3   # 0.3 std devs below average income
z_credit =  0.8   # 0.8 std devs above average credit score

print('Model: ŷ = w_age×z_age + w_income×z_income + w_credit×z_credit + b')
print()
print(f'= {w_age}×{z_age} + {w_income}×({z_income}) + {w_credit}×{z_credit} + {b}')

term_age    = w_age * z_age
term_income = w_income * z_income
term_credit = w_credit * z_credit
y_hat = term_age + term_income + term_credit + b

print(f'= {term_age:.2f} + ({term_income:.2f}) + {term_credit:.2f} + {b}')
print(f'= {y_hat:.2f}')
print()
decision = 'APPROVE' if y_hat >= 0.5 else 'REJECT'
print(f'Prediction: {y_hat:.2f} ({y_hat*100:.0f}% approval probability)')
print(f'Decision: {decision} (threshold = 0.5)')

# Visualize contribution of each feature
contributions = [term_age, term_income, term_credit, b]
labels = ['Age', 'Income', 'Credit', 'Bias']
colors = ['green' if c >= 0 else 'red' for c in contributions]

fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.bar(labels, contributions, color=colors, alpha=0.7, edgecolor='black')
ax.axhline(0, color='black', linewidth=0.8)
ax.set_title(f'Feature Contributions to Prediction\n(Total = {y_hat:.2f} → {decision})')
ax.set_ylabel('Contribution to ŷ')
for bar, val in zip(bars, contributions):
    ax.text(bar.get_x() + bar.get_width()/2, val + 0.01 * np.sign(val), f'{val:.2f}',
            ha='center', va='bottom' if val >= 0 else 'top', fontweight='bold')
plt.tight_layout()
plt.show()
````

````python
# --- Exercise 3.2: Interpreting Weights ---
weights = {'age': 0.4, 'income': 0.5, 'credit': 0.6}

print('=== INTERPRETING WEIGHTS ===')
print()
for feature, w in sorted(weights.items(), key=lambda x: abs(x[1]), reverse=True):
    direction = 'Higher = more likely approved' if w > 0 else 'Higher = less likely approved'
    importance = 'HIGH' if abs(w) > 0.5 else 'MEDIUM' if abs(w) > 0.3 else 'LOW'
    print(f'w_{feature} = {w}')
    print(f'  Direction:  {direction}')
    print(f'  Importance: {importance}')
    print(f'  Meaning: +1 std dev in {feature} adds {w} to prediction')
    print()

most_important = max(weights, key=lambda x: abs(weights[x]))
print(f'Most important feature: {most_important} (weight = {weights[most_important]})')
print('\nKEY INSIGHT: Larger |weight| = feature has more influence on prediction.')

fig, ax = plt.subplots(figsize=(7, 4))
features = list(weights.keys())
values = list(weights.values())
colors = ['steelblue' if v > 0 else 'salmon' for v in values]
ax.barh(features, values, color=colors, alpha=0.8, edgecolor='black')
ax.axvline(0, color='black', linewidth=0.8)
ax.set_xlabel('Weight Value')
ax.set_title('Feature Weights — Importance and Direction')
for i, (f, v) in enumerate(zip(features, values)):
    ax.text(v + 0.01, i, f'{v}', va='center', fontweight='bold')
plt.tight_layout()
plt.show()
````

# WEEK 4-5: How Learning Happens — Gradient Descent
## Concept: The Error Landscape

Think of error as a **hill**. We want to reach the bottom (minimum error).  
The **gradient** is the slope — it tells us which direction is downhill.

**Gradient formula for weight w:**
$$\frac{\partial L}{\partial w} = 2(\hat{y} - y) \times z$$

- Negative gradient → weight too small → increase it  
- Positive gradient → weight too large → decrease it  
- Zero gradient → we're at the bottom (converged!)

````python
# --- Visualize the Error Landscape ---
# For single weight w, one data point z=1.5, y=1
z, y_actual = 1.5, 1.0

w_range = np.linspace(-0.5, 1.5, 200)
errors = [(w * z - y_actual) ** 2 for w in w_range]

# Current position
w_current = 0.1
e_current = (w_current * z - y_actual) ** 2
grad_current = 2 * (w_current * z - y_actual) * z

print(f'At w = {w_current}:')
print(f'  Prediction: ŷ = {w_current} × {z} = {w_current * z}')
print(f'  Error: ({w_current * z} - {y_actual})² = {e_current:.4f}')
print(f'  Gradient: 2 × ({w_current * z:.2f} - {y_actual}) × {z} = {grad_current:.4f}')
print(f'  Gradient is negative → weight is too small → must INCREASE')

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(w_range, errors, 'b-', linewidth=2.5, label='Error (ŷ - y)²')
ax.scatter([w_current], [e_current], s=150, color='red', zorder=5, label=f'Current w = {w_current}')

# Draw tangent (gradient arrow)
tangent_x = np.array([w_current - 0.15, w_current + 0.15])
tangent_y = e_current + grad_current * (tangent_x - w_current)
ax.plot(tangent_x, tangent_y, 'r-', linewidth=2, label=f'Gradient = {grad_current:.2f} (slope)')
ax.annotate('', xy=(w_current - 0.25, e_current + 0.15), xytext=(w_current, e_current),
            arrowprops=dict(arrowstyle='->', color='green', lw=2))
ax.text(w_current - 0.35, e_current + 0.2, 'Move\nthis way!', color='green', fontsize=10)

# Mark minimum
w_opt = y_actual / z
e_opt = (w_opt * z - y_actual) ** 2
ax.scatter([w_opt], [e_opt], s=200, color='gold', zorder=5, marker='*', label=f'Optimal w = {w_opt:.2f}')

ax.set_xlabel('Weight (w)')
ax.set_ylabel('Error')
ax.set_title('The Error Landscape — We Want the Bottom!')
ax.legend()
plt.tight_layout()
plt.show()
````

````python
# --- Exercise 4.1: Computing Gradient ---
y_hat    = 0.15   # prediction
y_actual = 1.0    # actual label
z        = 1.5    # normalized feature

print('Given:')
print(f'  Prediction (ŷ) = {y_hat}')
print(f'  Actual (y)     = {y_actual}')
print(f'  Feature (z)    = {z}')
print()
print('Gradient Formula: ∂L/∂w = 2 × (ŷ - y) × z')
print()

error     = y_hat - y_actual
gradient  = 2 * error * z

print(f'Step 1: Prediction error = {y_hat} - {y_actual} = {error}')
print(f'Step 2: ∂L/∂w = 2 × {error} × {z} = {gradient}')
print()

if gradient < 0:
    print(f'Gradient = {gradient} (NEGATIVE)')
    print('  → Weight is TOO SMALL')
    print('  → We must INCREASE the weight')
    print(f'  → Magnitude {abs(gradient):.2f} is large → big room for improvement')
elif gradient > 0:
    print(f'Gradient = {gradient} (POSITIVE)')
    print('  → Weight is TOO LARGE')
    print('  → We must DECREASE the weight')
else:
    print('Gradient = 0 → Already at optimum! Stop learning.')
````

````python
# --- Exercise 4.2 & 4.3: Updating Weight & Verifying Improvement ---
w_old    = 0.1
gradient = -2.55
lr       = 0.1
z        = 1.5
y_actual = 1.0

print('Weight Update Formula: w_new = w_old - lr × gradient')
print()
w_new = w_old - lr * gradient
print(f'w_new = {w_old} - {lr} × ({gradient})')
print(f'w_new = {w_old} - ({lr * gradient})')
print(f'w_new = {w_new}')
print(f'Weight changed from {w_old} to {w_new} (increased by {w_new - w_old:.3f})')

# Verify error decreased
y_hat_old  = w_old * z
y_hat_new  = w_new * z
error_old  = (y_hat_old - y_actual) ** 2
error_new  = (y_hat_new - y_actual) ** 2
improvement = (1 - error_new / error_old) * 100

print()
print('Verification:')
print(f'  OLD: ŷ = {w_old} × {z} = {y_hat_old:.4f}  →  error = {error_old:.4f}')
print(f'  NEW: ŷ = {w_new:.3f} × {z} = {y_hat_new:.4f}  →  error = {error_new:.4f}')
print(f'  Error reduced by {improvement:.1f}% ✓  The update worked!')
````

## Concept: Multiple Iterations — Convergence

One update isn't enough. We repeat the cycle:
1. Predict  
2. Compute error  
3. Compute gradient  
4. Update weight  
5. Repeat

Each iteration, the error gets smaller and the gradient shrinks toward 0. That's **convergence**.

````python
# --- Exercise 4.4: Watch Convergence Happen ---
w = 0.1       # starting weight
z = 1.5       # feature
y = 1.0       # actual
lr = 0.1      # learning rate
n_iterations = 20

history = []
print(f'{'Iter':>4} | {'w':>8} | {'ŷ':>8} | {'Error':>8} | {'Gradient':>10}')
print('-' * 55)

for i in range(n_iterations):
    y_hat    = w * z
    error    = (y_hat - y) ** 2
    gradient = 2 * (y_hat - y) * z
    history.append((i+1, w, y_hat, error, gradient))
    
    if i < 7 or i == n_iterations - 1:
        print(f'{i+1:>4} | {w:>8.4f} | {y_hat:>8.4f} | {error:>8.4f} | {gradient:>10.4f}')
    elif i == 7:
        print('  ...')
    
    w = w - lr * gradient

print(f'\nFinal weight: {w:.4f} (optimal is {y/z:.4f})')
print('Notice: error decreases, gradient shrinks to ~0 — that is CONVERGENCE!')

# Visualize convergence
iters, ws, y_hats, errors, grads = zip(*history)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

axes[0].plot(iters, errors, 'r-o', markersize=4)
axes[0].set_title('Error Decreasing')
axes[0].set_xlabel('Iteration')
axes[0].set_ylabel('Squared Error')

axes[1].plot(iters, grads, 'b-o', markersize=4)
axes[1].axhline(0, color='black', linestyle='--', linewidth=0.8)
axes[1].set_title('Gradient Shrinking → 0')
axes[1].set_xlabel('Iteration')
axes[1].set_ylabel('Gradient')

axes[2].plot(iters, ws, 'g-o', markersize=4)
axes[2].axhline(y/z, color='gold', linestyle='--', linewidth=2, label=f'Optimal w = {y/z:.3f}')
axes[2].set_title('Weight Converging to Optimum')
axes[2].set_xlabel('Iteration')
axes[2].set_ylabel('Weight')
axes[2].legend()

plt.suptitle('Gradient Descent Convergence', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
````

# WEEK 6-7: Multiple Weights Learning Together
## Concept: Each Weight Has Its Own Gradient

For a model with multiple features, each weight gets its own gradient:  
$$\frac{\partial L}{\partial w_j} = 2(\hat{y} - y) \times z_j$$

Same error term, but **different features**. All weights update simultaneously.

````python
# --- Exercise 5.1: Multiple Weights, One Data Point ---
# Model: ŷ = w1×z1 + w2×z2 + b
z1, z2 = 1.5, 0.5  # features
y = 1.0             # actual
w1, w2, b = 0.1, 0.1, 0.0  # initial weights
lr = 0.1

print('=== MULTIPLE WEIGHTS UPDATE ===')
print(f'Model: ŷ = w1×z1 + w2×z2 + b')
print(f'Features: z1={z1}, z2={z2}')
print(f'Initial weights: w1={w1}, w2={w2}, b={b}')
print()

# Step 1: Predict
y_hat = w1 * z1 + w2 * z2 + b
print(f'Step 1 - Predict: ŷ = {w1}×{z1} + {w2}×{z2} + {b} = {y_hat}')

# Step 2: Error
error = y_hat - y
print(f'Step 2 - Error: ŷ - y = {y_hat} - {y} = {error}')

# Step 3: Gradients
grad_w1 = 2 * error * z1
grad_w2 = 2 * error * z2
print(f'Step 3 - Gradients (INDEPENDENT):')
print(f'  ∂L/∂w1 = 2 × {error} × {z1} = {grad_w1}')
print(f'  ∂L/∂w2 = 2 × {error} × {z2} = {grad_w2}')

# Step 4: Simultaneous updates
w1_new = w1 - lr * grad_w1
w2_new = w2 - lr * grad_w2
print(f'Step 4 - SIMULTANEOUS Updates:')
print(f'  w1_new = {w1} - {lr}×{grad_w1} = {w1_new}')
print(f'  w2_new = {w2} - {lr}×{grad_w2} = {w2_new}')
print()
print('KEY: Same error (-0.8), but DIFFERENT gradients:')
print(f'  z1={z1} → gradient {grad_w1} → bigger update')
print(f'  z2={z2} → gradient {grad_w2} → smaller update')
print('Features with larger z contribute more to the gradient!')
````

````python
# --- Exercise 5.2: Learning Rate Effects ---
gradient = -2.55
w_initial = 0.1

learning_rates = [0.01, 0.1, 0.5, 1.0]
labels = ['0.01 (too small)', '0.1 (just right)', '0.5 (risky)', '1.0 (diverges)']

print(f'Gradient = {gradient}, Initial w = {w_initial}')
print()
print(f'{'LR':>5} | {'w_new':>8} | {'Change':>8} | Assessment')
print('-' * 55)
for lr, label in zip(learning_rates, labels):
    w_new = w_initial - lr * gradient
    change = w_new - w_initial
    print(f'{lr:>5} | {w_new:>8.4f} | {change:>8.4f} | {label}')

# Visualize effects over 15 iterations with different LRs
z, y_actual = 1.5, 1.0
n_iters = 15

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
colors = ['blue', 'green', 'orange', 'red']

for lr, label, color in zip([0.01, 0.1, 0.5, 1.0], labels, colors):
    w = 0.1
    errors_over_time = []
    for _ in range(n_iters):
        y_hat = w * z
        error = (y_hat - y_actual) ** 2
        errors_over_time.append(error)
        grad = 2 * (y_hat - y_actual) * z
        w = w - lr * grad
        if abs(w) > 100:  # diverging
            errors_over_time += [float('nan')] * (n_iters - len(errors_over_time))
            break
    axes[0].plot(errors_over_time, label=f'lr={lr}', color=color, linewidth=2)

axes[0].set_title('Learning Rate Effect on Convergence')
axes[0].set_xlabel('Iteration')
axes[0].set_ylabel('Error')
axes[0].set_ylim(-0.1, 1.5)
axes[0].legend()

# Analogy visualization
hill_x = np.linspace(-1, 3, 200)
hill_y = (hill_x - 0.67) ** 2
axes[1].plot(hill_x, hill_y, 'b-', linewidth=2, label='Error landscape')
step_sizes = [0.05, 0.25, 0.7, 1.4]
start = 2.5
axes[1].scatter([start], [(start - 0.67)**2], s=200, color='black', zorder=5)
for step, lr_label, color in zip(step_sizes, ['Too small', 'Just right', 'Risky', 'Too large'], colors):
    new_x = start - step
    axes[1].annotate('', xy=(new_x, (new_x - 0.67)**2), xytext=(start, (start - 0.67)**2),
                     arrowprops=dict(arrowstyle='->', color=color, lw=2))
axes[1].set_title('Step Size Analogy')
axes[1].set_xlabel('Weight value')
axes[1].set_ylabel('Error')
axes[1].set_xticks([])

plt.suptitle('Learning Rate: Not Too Small, Not Too Large', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
````

# WEEK 8: The Complete Learning System
## Concept: The Full Training Loop

Real training uses **many data points**.  
For each epoch, we loop over all examples and average the gradients.

```
FOR each epoch:
    total_gradient = 0
    FOR each training example:
        1. Predict
        2. Compute error
        3. Accumulate gradient
    Average gradient across examples
    Update weight
    Check convergence
```

````python
# --- The Complete Training Loop from Scratch ---
np.random.seed(42)

# Synthetic dataset: predict loan approval from age + income
n_samples = 100
ages    = np.random.uniform(20, 70, n_samples)
incomes = np.random.uniform(20, 100, n_samples)

# True relationship: older + richer → approved
true_w_age = 0.5
true_w_inc = 0.7
true_b     = 0.0
y_true_raw = true_w_age * ages / 50 + true_w_inc * incomes / 100 + np.random.normal(0, 0.1, n_samples)
y_true = (y_true_raw > np.median(y_true_raw)).astype(float)

# Step 1: Normalize features
z_ages    = (ages    - ages.mean())    / ages.std()
z_incomes = (incomes - incomes.mean()) / incomes.std()
X = np.column_stack([z_ages, z_incomes])

# Step 2: Initialize weights
w = np.array([0.0, 0.0])
b = 0.0
lr = 0.05
n_epochs = 100

# Training loop
loss_history = []

for epoch in range(n_epochs):
    # Forward pass: predict for all examples
    y_hat = X @ w + b            # matrix multiply features by weights
    
    # Compute loss (MSE)
    loss = np.mean((y_hat - y_true) ** 2)
    loss_history.append(loss)
    
    # Backward pass: gradients (averaged over all examples)
    errors = y_hat - y_true
    grad_w = 2 * X.T @ errors / n_samples
    grad_b = 2 * np.mean(errors)
    
    # Update weights
    w -= lr * grad_w
    b -= lr * grad_b

print('Training complete!')
print(f'Final weights: w_age={w[0]:.3f}, w_income={w[1]:.3f}, bias={b:.3f}')
print(f'Final loss: {loss_history[-1]:.4f}')
print(f'Initial loss: {loss_history[0]:.4f}')
print(f'Improvement: {(1 - loss_history[-1]/loss_history[0])*100:.1f}%')

print(f'\nModel learned: income matters more than age')
print(f'  w_age = {w[0]:.3f}  (true ~ 0.5 normalized)')
print(f'  w_inc = {w[1]:.3f}  (true ~ 0.7 normalized)')

# Visualize training
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].plot(loss_history, 'b-', linewidth=2)
axes[0].set_title('Training Loss Decreasing')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE Loss')
axes[0].axhline(loss_history[-1], color='red', linestyle='--', label=f'Final = {loss_history[-1]:.3f}')
axes[0].legend()

# Decision boundary
xx, yy = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
zz = np.column_stack([xx.ravel(), yy.ravel()]) @ w + b
zz = zz.reshape(xx.shape)

axes[1].contourf(xx, yy, zz, levels=[zz.min(), 0.5, zz.max()], alpha=0.2, colors=['red', 'green'])
axes[1].contour(xx, yy, zz, levels=[0.5], colors='black', linewidths=2)
scatter = axes[1].scatter(z_ages, z_incomes, c=y_true, cmap='RdYlGn', s=30, alpha=0.7, edgecolors='none')
axes[1].set_xlabel('Z-Age')
axes[1].set_ylabel('Z-Income')
axes[1].set_title('Learned Decision Boundary')
plt.colorbar(scatter, ax=axes[1], label='Approved (1) / Rejected (0)')

plt.suptitle('Complete Training Loop: From Random Weights to Learned Model', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
````

# WEEK 9: Why Statistics Matter for Learning
## Concept: Normalization Enables Balanced Learning

This week we circle back to Week 1 with new eyes.  
Normalization isn't just cleaning — it **enables fair gradient descent**.

````python
# --- Side-by-side: Learning WITH vs WITHOUT Normalization ---
np.random.seed(42)
n = 100

# Raw features
ages_raw    = np.random.uniform(20, 70, n)          # range ~50
incomes_raw = np.random.uniform(20000, 100000, n)   # range ~80000 !
y = ((ages_raw / 70 + incomes_raw / 100000) > 1.0).astype(float)

def train_model(X_train, y_train, lr=0.001, n_epochs=200):
    w = np.zeros(X_train.shape[1])
    b = 0.0
    losses = []
    for _ in range(n_epochs):
        y_hat = X_train @ w + b
        loss = np.mean((y_hat - y_train)**2)
        losses.append(loss)
        errors = y_hat - y_train
        w -= lr * 2 * X_train.T @ errors / len(y_train)
        b -= lr * 2 * np.mean(errors)
    return w, b, losses

# WITHOUT normalization
X_raw = np.column_stack([ages_raw, incomes_raw])
w_raw, b_raw, losses_raw = train_model(X_raw, y, lr=1e-9, n_epochs=200)

# WITH normalization
X_norm = np.column_stack([
    (ages_raw - ages_raw.mean()) / ages_raw.std(),
    (incomes_raw - incomes_raw.mean()) / incomes_raw.std()
])
w_norm, b_norm, losses_norm = train_model(X_norm, y, lr=0.1, n_epochs=200)

print('WITHOUT normalization:')
print(f'  w_age = {w_raw[0]:.6f}  (contribution barely visible)')
print(f'  w_inc = {w_raw[1]:.6f}  (tiny because income values are huge)')
print(f'  Final loss: {losses_raw[-1]:.4f}')
print()
print('WITH normalization:')
print(f'  w_age = {w_norm[0]:.4f}  (meaningful weight)')
print(f'  w_inc = {w_norm[1]:.4f}  (meaningful weight)')
print(f'  Final loss: {losses_norm[-1]:.4f}')
print()
print('Normalization allows fair, balanced learning!')

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(losses_raw, 'r-', linewidth=2, label='Without normalization')
axes[0].set_title('Without Normalization: Slow / Stuck')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE Loss')
axes[0].legend()

axes[1].plot(losses_norm, 'g-', linewidth=2, label='With normalization')
axes[1].set_title('With Normalization: Fast Convergence')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('MSE Loss')
axes[1].legend()

plt.suptitle('Why Z-Score Normalization is NOT Optional', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
````

# WEEK 10-11: Complex Models
## Concept: When Linear Isn't Enough

Linear models can only fit **straight lines** (or planes in higher dimensions).  
Real data often has **curves**.

**Two solutions:**
1. **Polynomial features:** manually create z², z³, z₁×z₂
2. **Decision trees:** automatically find splits in the data

````python
# --- Linear vs Polynomial: Fitting Curves ---
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

np.random.seed(42)
x = np.linspace(-3, 3, 100)
y_nonlinear = np.sin(x) + 0.2 * np.random.randn(100)  # curved relationship

X = x.reshape(-1, 1)

# Fit models of different complexity
models = {
    'Linear (degree 1)':    make_pipeline(PolynomialFeatures(1), LinearRegression()),
    'Polynomial (degree 3)': make_pipeline(PolynomialFeatures(3), LinearRegression()),
    'Polynomial (degree 9)': make_pipeline(PolynomialFeatures(9), LinearRegression()),
}

print('Model          | Train MSE | Parameters')
print('-' * 42)
x_plot = np.linspace(-3, 3, 300).reshape(-1, 1)

fig, axes = plt.subplots(1, 3, figsize=(14, 4), sharey=True)
colors = ['blue', 'green', 'red']

for ax, (name, model), color in zip(axes, models.items(), colors):
    model.fit(X, y_nonlinear)
    y_pred = model.predict(X)
    y_plot = model.predict(x_plot)
    mse = np.mean((y_pred - y_nonlinear)**2)
    degree = model.steps[0][1].degree
    n_params = degree + 1
    print(f'{name:30} | {mse:.5f}   | {n_params}')
    
    ax.scatter(x, y_nonlinear, alpha=0.4, color='gray', s=20, label='Data')
    ax.plot(x_plot, y_plot, color=color, linewidth=2.5, label=f'MSE={mse:.3f}')
    ax.set_title(name)
    ax.set_xlabel('x')
    ax.legend(fontsize=8)
    ax.set_ylim(-2, 2)

axes[0].set_ylabel('y')
plt.suptitle('Model Complexity: Underfitting → Good Fit → Overfitting Risk', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
print('\nHigh-degree polynomial fits training data perfectly — but will it generalize?')
````

````python
# --- The Overfitting Problem: Train vs Test Error ---
from sklearn.model_selection import train_test_split

np.random.seed(0)
x = np.linspace(-3, 3, 80)
y_data = np.sin(x) + 0.3 * np.random.randn(80)

X = x.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y_data, test_size=0.3, random_state=42)

degrees = range(1, 12)
train_errors, test_errors = [], []

for d in degrees:
    model = make_pipeline(PolynomialFeatures(d), LinearRegression())
    model.fit(X_train, y_train)
    train_errors.append(np.mean((model.predict(X_train) - y_train)**2))
    test_errors.append(np.mean((model.predict(X_test)  - y_test)**2))

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(degrees, train_errors, 'b-o', label='Training Error', linewidth=2)
ax.plot(degrees, test_errors,  'r-o', label='Test Error',     linewidth=2)
ax.axvline(3, color='green', linestyle='--', linewidth=1.5, label='Sweet spot (degree 3)')
ax.set_xlabel('Polynomial Degree (Model Complexity)')
ax.set_ylabel('MSE')
ax.set_title('Bias-Variance Tradeoff: Training Error Always Drops,\nTest Error Has a Sweet Spot')
ax.legend()
ax.set_ylim(0, 0.6)
ax.annotate('Underfitting\n(too simple)', xy=(1, test_errors[0]), xytext=(1.5, 0.5),
            arrowprops=dict(arrowstyle='->', color='gray'))
ax.annotate('Overfitting\n(too complex)', xy=(9, test_errors[8]), xytext=(8.5, 0.5),
            arrowprops=dict(arrowstyle='->', color='gray'))
plt.tight_layout()
plt.show()
print('\nKEY INSIGHT: Choose model complexity by TEST error, not training error!')
````

# WEEK 12: Neural Networks
## Concept: Hidden Layers = Automatic Feature Engineering

A neural network is just a **stack of linear models** with **non-linearities** (activation functions) between them.

```
Input:   z_age, z_income, z_credit
         ↓ (weights + bias)
Hidden:  h1 = ReLU(w·z + b)   ← learned features
         ↓ (more weights)
Output:  ŷ = w·h + b
```

Same gradient descent, same weight update formula — just more layers.

````python
# --- ReLU Activation Function ---
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-4, 4, 200)

fig, axes = plt.subplots(1, 3, figsize=(13, 4))

axes[0].plot(x, x, 'b-', linewidth=2, label='Linear (no activation)')
axes[0].set_title('Linear: No Non-Linearity')
axes[0].set_ylabel('Output')
axes[0].axhline(0, color='black', linewidth=0.5)
axes[0].axvline(0, color='black', linewidth=0.5)
axes[0].legend()

axes[1].plot(x, relu(x), 'g-', linewidth=2.5, label='ReLU(x) = max(0, x)')
axes[1].set_title('ReLU: Negative → 0, Positive → Keeps')
axes[1].axhline(0, color='black', linewidth=0.5)
axes[1].axvline(0, color='black', linewidth=0.5)
axes[1].legend()

axes[2].plot(x, sigmoid(x), 'r-', linewidth=2.5, label='Sigmoid(x) = 1/(1+e⁻ˣ)')
axes[2].axhline(0.5, color='black', linestyle='--', linewidth=0.8)
axes[2].set_title('Sigmoid: Squashes to (0, 1)')
axes[2].axhline(0, color='black', linewidth=0.5)
axes[2].axvline(0, color='black', linewidth=0.5)
axes[2].legend()

for ax in axes:
    ax.set_xlabel('Input')
    ax.set_ylim(-1, 4)

plt.suptitle('Activation Functions: Adding Non-Linearity', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

# Demonstrate ReLU values
print('ReLU Examples:')
for val in [-2.0, -0.5, 0.0, 0.5, 2.0]:
    print(f'  ReLU({val:.1f}) = {relu(val):.1f}  →  {"passed through" if val > 0 else "blocked (→ 0)"}')
````

````python
# --- Neural Network vs Linear Model: Power of Hidden Layers ---
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression

np.random.seed(42)

# XOR-like non-linear problem that a linear model CANNOT solve
n = 200
X_xor = np.random.randn(n, 2)
y_xor = ((X_xor[:, 0] * X_xor[:, 1]) > 0).astype(int)

# Fit both models
linear = LogisticRegression()
linear.fit(X_xor, y_xor)
linear_acc = linear.score(X_xor, y_xor)

neural = MLPClassifier(hidden_layer_sizes=(16, 8), max_iter=500, random_state=42)
neural.fit(X_xor, y_xor)
neural_acc = neural.score(X_xor, y_xor)

print(f'Linear Model accuracy: {linear_acc:.1%}')
print(f'Neural Network accuracy: {neural_acc:.1%}')
print('\nNeural network learns the non-linear boundary — linear model cannot!')

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
xx, yy = np.meshgrid(np.linspace(-3, 3, 200), np.linspace(-3, 3, 200))
Xgrid = np.column_stack([xx.ravel(), yy.ravel()])

for ax, model, title in zip(axes, [linear, neural], 
                             [f'Linear Model\n({linear_acc:.1%} accuracy)',
                              f'Neural Network\n({neural_acc:.1%} accuracy)']):
    zz = model.predict(Xgrid).reshape(xx.shape)
    ax.contourf(xx, yy, zz, alpha=0.2, cmap='RdYlGn')
    ax.scatter(X_xor[:, 0], X_xor[:, 1], c=y_xor, cmap='RdYlGn', s=25, alpha=0.7, edgecolors='none')
    ax.set_title(title)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')

plt.suptitle('Why Hidden Layers Matter: Non-Linear Boundaries', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
````

# WEEK 13-14: Practical Mastery
## Concept: Overfitting, Evaluation, and Model Selection

The most important skill: knowing whether your model **actually works** on new data.

**Golden rules:**
- Train on training data only
- Never touch test data until final evaluation
- Large train-test gap = overfitting
- Choose model by test accuracy, not training accuracy

````python
# --- Exercise 8.1: Proper Train/Test Evaluation ---
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

np.random.seed(42)

X, y = make_classification(n_samples=300, n_features=10, n_informative=5, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)

models_to_compare = {
    'Model A — Linear':           LogisticRegression(random_state=42),
    'Model B — Decision Tree':    DecisionTreeClassifier(max_depth=None, random_state=42),
    'Model C — Neural Network':   MLPClassifier(hidden_layer_sizes=(50, 25), max_iter=500, random_state=42),
}

print(f'{'Model':35} | {'Train Acc':10} | {'Test Acc':10} | {'Gap':8} | Assessment')
print('-' * 90)

results = {}
for name, model in models_to_compare.items():
    model.fit(X_train, y_train)
    train_acc = model.score(X_train, y_train)
    test_acc  = model.score(X_test,  y_test)
    gap = train_acc - test_acc
    results[name] = (train_acc, test_acc, gap)
    
    if gap > 0.15:
        assessment = 'OVERFIT ⚠️'
    elif gap > 0.08:
        assessment = 'Some overfitting'
    else:
        assessment = 'Good generalization ✓'
    
    print(f'{name:35} | {train_acc:10.1%} | {test_acc:10.1%} | {gap:8.1%} | {assessment}')

best = max(results, key=lambda k: results[k][1])
print(f'\nBest model by test accuracy: {best}')
print('RULE: Never choose model by training accuracy. Always use test accuracy!')
````

````python
# --- Exercise 8.2: Visualize Overfitting ---
model_names = list(results.keys())
train_accs  = [results[m][0] for m in model_names]
test_accs   = [results[m][1] for m in model_names]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

x = np.arange(len(model_names))
width = 0.35
axes[0].bar(x - width/2, train_accs, width, label='Train Accuracy', color='steelblue', alpha=0.8)
axes[0].bar(x + width/2, test_accs,  width, label='Test Accuracy',  color='salmon', alpha=0.8)
axes[0].set_xticks(x)
axes[0].set_xticklabels(['Linear', 'Tree', 'Neural Net'], fontsize=10)
axes[0].set_ylabel('Accuracy')
axes[0].set_title('Train vs Test Accuracy by Model')
axes[0].set_ylim(0.5, 1.1)
axes[0].legend()
for xi, (tr, te) in zip(x, zip(train_accs, test_accs)):
    axes[0].text(xi - width/2, tr + 0.01, f'{tr:.1%}', ha='center', fontsize=8)
    axes[0].text(xi + width/2, te + 0.01, f'{te:.1%}', ha='center', fontsize=8)

# Gap visualization
gaps = [results[m][2] for m in model_names]
bar_colors = ['green' if g <= 0.08 else 'orange' if g <= 0.15 else 'red' for g in gaps]
axes[1].bar(['Linear', 'Tree', 'Neural Net'], gaps, color=bar_colors, alpha=0.8, edgecolor='black')
axes[1].axhline(0.08, color='orange', linestyle='--', label='Overfitting threshold')
axes[1].set_ylabel('Train - Test Gap')
axes[1].set_title('Generalization Gap (Smaller = Better)')
axes[1].legend()
for i, g in enumerate(gaps):
    axes[1].text(i, g + 0.005, f'{g:.1%}', ha='center', fontweight='bold')

plt.suptitle('Model Evaluation: Finding the Generalization Sweet Spot', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
````

## Concept: Regularization — Preventing Overfitting

Regularization **penalizes large weights**, forcing the model to stay simple.

**L2 Regularization (Ridge):** add λ×Σw² to the loss  
**L1 Regularization (Lasso):** add λ×Σ|w| to the loss (makes some weights exactly 0)

````python
# --- Regularization in Action ---
from sklearn.linear_model import Ridge, Lasso

# Overfit scenario: high-degree polynomial on small data
np.random.seed(10)
x_small = np.linspace(-1, 1, 15)
y_small = np.sin(np.pi * x_small) + 0.2 * np.random.randn(15)
x_plot  = np.linspace(-1, 1, 300)

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=10)
X_small_poly = poly.fit_transform(x_small.reshape(-1, 1))
X_plot_poly  = poly.transform(x_plot.reshape(-1, 1))

configs = [
    ('No regularization\n(overfits)', Ridge(alpha=0.0001), 'red'),
    ('L2 Regularization\n(Ridge, α=1)', Ridge(alpha=1.0), 'blue'),
    ('L2 Regularization\n(Ridge, α=10)', Ridge(alpha=10.0), 'green'),
]

fig, axes = plt.subplots(1, 3, figsize=(14, 4), sharey=True)
true_y = np.sin(np.pi * x_plot)

for ax, (title, model, color) in zip(axes, configs):
    model.fit(X_small_poly, y_small)
    y_pred = model.predict(X_plot_poly)
    
    ax.scatter(x_small, y_small, color='black', s=40, zorder=5, label='Training data')
    ax.plot(x_plot, true_y, 'k--', linewidth=1.5, label='True curve', alpha=0.5)
    ax.plot(x_plot, y_pred, color=color, linewidth=2.5, label='Model')
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylim(-2.5, 2.5)
    ax.legend(fontsize=7)

axes[0].set_ylabel('y')
plt.suptitle('Regularization: Taming Overfitting by Penalizing Large Weights', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
print('Appropriate regularization → model fits the TRUE pattern, not the noise!')
````

## Putting Everything Together

Now we apply the full pipeline on a real dataset:  
**Load → EDA → Normalize → Split → Train → Evaluate → Compare**

````python
# --- Full ML Pipeline: Breast Cancer Classification ---
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

# 1. LOAD DATA
data = load_breast_cancer()
X_raw, y = data.data, data.target
feature_names = data.feature_names
print('=== STEP 1: DATA EXPLORATION ===')
print(f'Samples: {X_raw.shape[0]}, Features: {X_raw.shape[1]}')
print(f'Classes: {data.target_names}')
print(f'Class balance: {y.mean():.1%} benign ({(y==0).sum()} malignant, {(y==1).sum()} benign)')
print()

# 2. STATISTICS ANALYSIS (applying Week 1 knowledge!)
print('=== STEP 2: STATISTICS (applying Week 1) ===')
print(f'{'Feature':35} | {'Mean':10} | {'Std Dev':10} | {'Range'}')
print('-' * 75)
for i, fname in enumerate(feature_names[:5]):
    col = X_raw[:, i]
    print(f'{fname:35} | {col.mean():10.2f} | {col.std():10.2f} | [{col.min():.2f}, {col.max():.2f}]')
print('  ... (all 30 features)')
print()

# 3. NORMALIZE (applying Week 1 z-score knowledge)
print('=== STEP 3: NORMALIZE ===' )
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X_raw)
print(f'Before: Mean range = [{X_raw.mean(axis=0).min():.2f}, {X_raw.mean(axis=0).max():.2f}]')
print(f'After:  Mean range = [{X_normalized.mean(axis=0).min():.2f}, {X_normalized.mean(axis=0).max():.2f}]')
print(f'After:  Std range  = [{X_normalized.std(axis=0).min():.2f}, {X_normalized.std(axis=0).max():.2f}]')
print()

# 4. SPLIT (applying Week 13-14 knowledge)
print('=== STEP 4: TRAIN/TEST SPLIT ===')
X_tr, X_te, y_tr, y_te = train_test_split(X_normalized, y, test_size=0.2, random_state=42)
print(f'Training: {len(X_tr)} samples | Test: {len(X_te)} samples')
print()

# 5. TRAIN & EVALUATE MULTIPLE MODELS
print('=== STEP 5: TRAIN & EVALUATE ===')
all_models = {
    'Linear (Logistic Regression)':  LogisticRegression(random_state=42, max_iter=1000),
    'Decision Tree (depth=3)':       DecisionTreeClassifier(max_depth=3, random_state=42),
    'Decision Tree (full depth)':    DecisionTreeClassifier(max_depth=None, random_state=42),
    'Neural Network':                MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42),
}

print(f'{'Model':40} | {'Train':8} | {'Test':8} | {'Gap':6} | {'CV 5-fold':10}')
print('-' * 90)
final_results = {}
for name, model in all_models.items():
    model.fit(X_tr, y_tr)
    tr_acc = model.score(X_tr, y_tr)
    te_acc = model.score(X_te, y_te)
    cv     = cross_val_score(model, X_normalized, y, cv=5).mean()
    final_results[name] = (tr_acc, te_acc, cv)
    star = ' ← BEST' if te_acc == max(final_results[n][1] for n in final_results) else ''
    print(f'{name:40} | {tr_acc:8.1%} | {te_acc:8.1%} | {tr_acc-te_acc:6.1%} | {cv:10.1%}{star}')

best_model = max(final_results, key=lambda k: final_results[k][1])
print(f'\n=== STEP 6: FINAL RECOMMENDATION ===')
print(f'Best model: {best_model}')
print(f'Test accuracy: {final_results[best_model][1]:.1%}')
print(f'\nThis model generalizes best to unseen data.')
````

````python
# --- Visualize: Feature Importance from Linear Model ---
lin_model = all_models['Linear (Logistic Regression)']
importances = np.abs(lin_model.coef_[0])
top_k = 10
top_idx = np.argsort(importances)[-top_k:][::-1]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Feature importances
axes[0].barh(range(top_k), importances[top_idx], color='steelblue', alpha=0.8)
axes[0].set_yticks(range(top_k))
axes[0].set_yticklabels([feature_names[i] for i in top_idx], fontsize=8)
axes[0].set_xlabel('|Weight|')
axes[0].set_title('Top 10 Most Important Features\n(Linear Model Weights)')

# Final comparison
model_labels = [n.split('(')[0].strip() for n in final_results.keys()]
te_accs = [final_results[n][1] for n in final_results]
cv_accs = [final_results[n][2] for n in final_results]

x = np.arange(len(model_labels))
width = 0.35
axes[1].bar(x - width/2, te_accs, width, label='Test Accuracy', color='salmon', alpha=0.8)
axes[1].bar(x + width/2, cv_accs, width, label='CV Accuracy', color='steelblue', alpha=0.8)
axes[1].set_xticks(x)
axes[1].set_xticklabels(model_labels, fontsize=8, rotation=10, ha='right')
axes[1].set_ylabel('Accuracy')
axes[1].set_title('Final Model Comparison')
axes[1].set_ylim(0.85, 1.0)
axes[1].legend()

plt.suptitle('Capstone: Full Analysis of Breast Cancer Dataset', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
print('\nWe applied EVERY concept from the 14-week roadmap in this single pipeline!')
````

## Common Misconceptions — Final Reminders

| Misconception | Reality |
|---|---|
| "Statistics are for ML" | Statistics *describe*. Learning *predicts*. But stats *enable* learning (normalization!). |
| "Complex models are always better" | Complex models overfit. Simple models generalize. Test error is truth. |
| "Gradient descent finds global optimum" | For linear models: yes. For deep nets: local optimum — usually good enough! |
| "Neural networks are magic" | Same principle as linear models. More layers + activations + backprop. No magic. |
| "Learning rate doesn't matter much" | Critical parameter. Too small = slow. Too large = diverges. Tune it. |
| "Training accuracy tells you if it works" | Only test accuracy tells you if it generalizes to real data. |
